```python
path='/Users/sakshamgarg/Desktop/software/jiit-hackathon-2018-master/own_train'
list=['c0','c1','c2','c3','c4','c5','c6','c7','c8','c9']
x=[]
y=[]
for i in list:
    #print(i)
    for root,dirs,files in os.walk(path+'/'+i):
        x=x+files

        y.append(i)
         
y1=[]
for i in y:
    for j in range(500):
        y1.append(i)
```

```python
# img_height=32
# img_width=32
# channels=3
# if (K.image_data_format()=='channels_last'):
#     input_shape=(len(train_data),img_height,img_width,channels)
# else:
#     input_shape=(len(train_data),channels,img_height,img_width)
#x_train=np.ndarray(shape=(5000,640,480,3),dtype=np.float32)
j=0
y_train=[]
x_train=[] 
for i,k in zip(df['img'],df['y']): 
    img=Image.open(path+"/"+k+'/'+i)
    #print(img)
    img=img.resize((224,224),PIL.Image.ANTIALIAS)
    #print(type(img))
    
    x=img_to_array(img)
    x/=255
    print(x.shape," ",j)
    #x=(x-128.0)/128.0
    x_train.append(x) 
    y_train.append(k)
    j+=1

```

```python
x_train=np.array(x_train)
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
label=LabelEncoder()
onehot=OneHotEncoder()
```

```python
a1=xtraintemp[:4500]
```

```python
b1=ytrain[:4500]
```

```python
from keras.models import Sequential
from keras.layers.core import Flatten, Dense, Dropout, Lambda
from keras.layers import BatchNormalization
from keras.optimizers import SGD, RMSprop, Adam
lr = 0.001
statefarm_model = Sequential()
statefarm_model.add(Dense(512, activation='relu', input_shape=(150528,)))
statefarm_model.add(BatchNormalization())
statefarm_model.add(Dense(10, activation='softmax'))
statefarm_model.compile(optimizer=Adam(lr=lr), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
statefarm_model.fit(a1,b1,epochs=10)
```

```python
xtesttemp=xtest.reshape(xtest.shape[0],xtest.shape[1]*xtest.shape[2]*xtest.shape[3])

```

```python
ypred=statefarm_model.predict(xtesttemp)
```

```python
ypred=np.argmax(ypred,axis=1)
```
